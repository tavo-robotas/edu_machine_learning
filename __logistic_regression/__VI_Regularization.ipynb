## Overfitting

By now we have seen we are familiar with couple different learning algoritmhs thats <code>**linear and logistic regression**</code> They work well for many problems but when we apply them to certain machine learning applications they can run into a problem called <code>**overfitting**</code> than can cause algorithm to perform very poorly on out of sample data. 

<code>**In linear regression:**</code>
    
Predicting housing prices with linear regression where we want to predict the price as a function of the size of the house.
 

 
   * One thing we can do we can fit a linear function to our dataset and we will get straight line. And that is not very good model. It seems pretty clear that as the size of the housing increase the housing prces flatten as we move to the right. This algoritmh doesn't fit the training set and we call this problem <code>**underfitting**</code> or <code>**high bias**</code>. Both of these mean that its just not even fitting the training data very well. <code>**Bias**</code> means that algorithm<code>**has a very strong preconception that housing prices are going to vary linearly with their size feature**</code> despite the data that shows the contrary, despite the evidence. It ends up being a poor fit to the data.
   
    
   * We can fit a quadratic function to the data and we get a curve that works pretty well. And thats is <code>**just right**</code>
   
    
   * The other extreme would be if we were to fit a fourth degree polynomial to the data. We have five  parameters. And with that we can actually fit a curve that passes through all five our training examples.And this curve on one hand seems to do a very good job fitting the the training set  but this is still a very wiggly curve going up and down all over the place and we dont actually think that it is a good model for predicting housing prices. This problem is called <code>**overfitting**</code> or <code>**high variance**</code>. Intuition is tha if we are fitting such a high order polynomial then the hypothesis can fit almost any function and <code>**space of possible hypothesis is just to large, to variable**</code>. And we dont have enough data to contsrain it to give us a good hypothesis so thats is called overfitting.
   
<code>**Outline:**</code><br>
If we have too many features, the learned <code>**hypothesis may fit the training set very well**</code> the cost function might be very close to zero:

\begin{multline*}
J(\theta) = \dfrac{1}{2m}\sum_{i=1}^m(h_{\theta}(x^{(i)} - y^{(i)})^2 \approx 0
\end{multline*}
    
But tries to hard to fit the training set so that it <code>**fails to generalize**</code> to new examples meaning failure predicting y on new examples. Term generalize refers to how well a hypothesis applies even to new examples to data that it has not seen.

<code>**In logistic regression:**</code><br>

A similar thing can apply to logistic regression as well. Lets say we have data set example <code>**with two features x1 and x2**</code>. 



  * We fit logistic regression with simple hypothesis where as usual <code>**g is our sigmoid function**</code> and we end up 
with a hypothesis trying to use like a straight line to seprate the positive and negative examples. And it doesn't look like a very good fit to the hypothesis. So once again this is an example of <code>**underfitting or high bias**</code>.

 * If we add to our features some <code>**quadratic terms**</code> then you can get a decision boundary that might look like quite good fit to the data. Maybe it is as good as we could get on this training set. 
 
 * At the other extreme if we were to fit a very high order polynomial, if we would generate lots of  <code>**high order polynomials terms**</code> of features then logistic regression may contort itself and try really hard to find a decision boundary that fits your training data and go to great lengths to contort itself to fit every single training example well. And it ends up being not very good hypothesis for making predictions.And this is an instance of <code>**overfitting or high variance. Its unlikely to generalize well to new examples.**</code>

When we will talk about <code>**debugging and diagnosing**</code> things that can go wrong with learning algorithms we will have specific tools to recognize when overfitting and also underfitting may be occuring. 



Plot hypothesis and see whats going on and select the approriate degree polynomial. Earlier for the housing prices example we could just plot the hypothesis and maybe see that it was fitting very wiggly function that goes all over the place to predict housing prices. And we could then use these figures like these to select an <code>**approriate degree polynomial**</code>
But it doesn't always work this way and if fact more often we may have learning <code>**problems that we just have a lot of features.**</code> And there is not just a matter of selecting what degree polynomial when we have so many features it also it becomes much harder to plot the data, to visualize it to decide what features to keep or not, which of them is usefull and useless.

## Regularization:



Main intuition on how regularization works with cost function. <code>**Suppose we penalize and make 3, 4 really small**</code>

## Our optimization objective

<code>**Usual squared error cost function:**</code>
<br>

Lets take our objective and modify it and add to it plus 1000 3 squared plus 1000 4 squared. 
\begin{multline*}
\min_{\theta} \dfrac{1}{2m} \sum_{i=1}^m(h_{\theta}(x^{(i)} - y^{(i)})^2 + 1000 \ \theta_{3}^{2} + 1000 \ \theta_{4}^{2} 
\end{multline*}

If we were to minimize this function, the only way to make this new cost function small is if 3 and 4 are small, othewrise if we have 1000 times 3 and 1000 times 4 our cost function is going to be huge. So when we minimize this new function we are going end up with  <code>**3 , 4  0**</code>. And thats as if we are getting rid of of these two last terms entirely and we are basically <code>**left with a quadratic function plus tiny contributions from small terms 3, 4 that may be very close to 0.**</code><br>

So far we looked into penalizing two of parameters values being large. More generaly the idea behind regularization.<br>

<code>**Regularization:**</code><br>
Smalls values for parameters, 0, 1, ..., n
 * "Simpler" hypothesis
 * Less prone to overfitting
 
<code>**Example:**</code>
Housing price prediction
 * Features: 1, 2, ..., 100
 * Parameters: 0, 1, 2, ... 100
 
And unlike polynomial example we dont know right away that 3, 4 are the high order polynomial terms.We have just a set of a hundred features its hard to pick in advance which are the ones that are less likely to be relevant or those parameters to pick to try to shrink them. So in regularization <code>**we augment our cost function (linear regression) with extra (regularization) term that will shrink every single parameter, except 0 we are not going to penalize it for being large**</code><br>

\begin{multline*}
J(\theta) = \dfrac{1}{2m} \ [\sum_{i=1}^m(h_{\theta}(x^{(i)}) - y^{(i)})^2 + \lambda \sum_{i=1}^n \theta_{j}^2]
\end{multline*}

<code>****</code> is called regularization parameter. And it control a trade of between two different goals. The first goal is that we would like to fit the training data well. The second goal is we want to keep the parameters small and thats captured by second term, by the regularizaton term. <code>**The lambda (regularization parameter)**</code> controls the trade of bewteen these two goals thus it keeps the hypothesis relatively simple to avoid overfitting. Thus if we still have all of the parameters we get a simple, smoother function that looks like but not entirely like quadratic function. And it gives much better hypothesis for this data.

### Gradient descent

For linear regression we have previously worked out two learning algorithms. One based on <code>**gradient descent**</code> and one based on the <code>**normal equation**</code>. We will take those two algorithms and generalize them to the case of regularized linear regression.<br>

\begin{multline*}
\dfrac{\lambda}{2m} \sum_{i=1}^n \theta_{j}^2
\end{multline*}

And we like to find parameters  that minimizes this cost function 
\begin{multline*}
\min_{\theta}J(\theta)
\end{multline*}


Previously we were using gradient descent for the original cost function without the regularization term. And let's write update for 0 seperatelly and the reason we have to do this is for our regularized linear regression we penelize all parameters except 0<br>

## Regularized logistic regression:

We previously talked about two types of optimization algorithms, thats gradient descent to optimize () and we also talked briefly about advanced otpimization methods. Ones that require that we provide a way to compute cost function () and that we provde a way to compute the derivatives. We can adapt both of those tecnhniques both gradient descent and the more advanced optimization techniques in order to have them work for regularized logistic regression. 



Earlier we saw that logistic regression can be prone to overfitting if we fit very high order polynomial features like this <code>**(g is sigmoid function)**</code>
\begin{multline*}
h_{\theta}(x) = g(\theta_{0} + \theta_{1}x_{1} + \theta_{2}x_{1}^{2} + \theta_{3}x_{1}^{2}x_{2} + \theta_{4}x_{1}^{2}x_{2}^{2} + \theta_{5}x_{1}^{2}x_{2}^{3} + \ ... )
\end{multline*}

In particulalr case we may end up where hypypothesis, its decision boundary  is sort of an overly complex, extremely contortive function that really isn't such a great hypothesis for this training set. And more generally if we have logistic regression with a lot of features not necessarily polynomial ones but just with a lot of features you can end up with overfitting.

 This was our cost function for logistic regression.

\begin{multline*}
J(\theta) = -\dfrac{1}{m}[ \sum_{i=1}^{m} y^{(i)}\log(h_{\theta}(x^{(i)})) - (1 -y^{(i)})\log(1 - h_{\theta}(x^{(i)}))]
\end{multline*}

To use regularization all we need to do is add to it the following term:
\begin{multline*}
+ \dfrac{\lambda}{2m} \sum_{j=1}^n \theta_{j}^{2}
\end{multline*}


\begin{multline*}
J(\theta) = -\dfrac{1}{m}[ \sum_{i=1}^{m} y^{(i)}\log(h_{\theta}(x^{(i)})) + (1 -y^{(i)})\log(1 - h_{\theta}(x^{(i)}))] + \dfrac{\lambda}{2m} \sum_{j=1}^n \theta_{j}^{2}
\end{multline*}

And the effect therefore is penalizing the parameters 1, 2, ... n for being too large.
If we do this then it will have the effect that even though we are fitting a very high order polynomial with a lot of features, but as long we apply regularization and keep the parameters small we are more likely to get a decision boundary. That maybe looks more like this



It looks more reasonable for separating the positive and negative examples. 

Implementing it on gradient descent look a lot like the previus one for linear regression

`repeat until convergence (simultaneously update all j){`
\begin{multline*}
\theta_{0} := \theta_{0} - \alpha \dfrac{1}{m} \sum_{i=1}^m(h_{\theta}(x^{(i)}) - y^{(i)})x_{0}^{(i)}
\end{multline*}
--------------------------------
\begin{multline*}
\theta_{j} := \theta_{j} - \alpha \dfrac{1}{m} \sum_{i=1}^m(h_{\theta}(x^{(i)}) - y^{(i)})x_{j}^{(i)}
\end{multline*}
\begin{multline*}
(j = 1, 2, 3, ..., n)
\end{multline*}
`}`

Modifying this algorithm to use regularized cost function all we need to do is pretty similar what with did with linear regression, just modify the second update rule as follows.

\begin{multline*}
\theta_{j} := \theta_{j} - \alpha [\dfrac{1}{m} \sum_{i=1}^m(h_{\theta}(x^{(i)}) - y^{(i)})x_{j}^{(i)} + \dfrac{\lambda}{m}\theta_{j}]
\end{multline*}

And cosmetically this look indetical what we had for linear regression but of course this is not the same algoritmh because now the hypothesis is defined using sigmoid function:

\begin{multline*}
h_{\theta}(x) = \dfrac{1}{1 + e^{-\theta^{\top}x}}
\end{multline*}

And these new regularized logistic regression term is of course the new partial deravative with respect to () of new cost function () that we defined earlier thad does use regularization

\begin{multline*}
\dfrac{\partial}{\partial\theta_{j}} J(\theta) = [\dfrac{1}{m} \sum_{i=1}^m(h_{\theta}(x^{(i)}) - y^{(i)})x_{j}^{(i)} + \dfrac{\lambda}{m}\theta_{j}]
\end{multline*}

\begin{multline*}
J(\theta) = -\dfrac{1}{m}[ \sum_{i=1}^{m} y^{(i)}\log(h_{\theta}(x^{(i)})) + (1 -y^{(i)})\log(1 - h_{\theta}(x^{(i)}))] + \dfrac{\lambda}{2m} \sum_{j=1}^n \theta_{j}^{2}
\end{multline*}

### Cross validation

Let's use cross validation to find the omptimal lambdas () for different models. To recap the main usage of cross validation is to tune the hypeparameters. The secondary objective is to estimate the mse of the test set. But when we do have the test set and we want to compare the models we simply compare the rmse for example ridge regression in the test set with rmse of the lasso in the test set. The main use case for cross validation is first tunning the hyper parameter and then estimating the performance metrics in the test set. If we do have the test set and if it's labeled that our best case when we want to compare models. However when we do cross validation when we are tunning hyperparametrs we want to tune them of the model on a data set thats the model is not going to see it for the test set. For that reason we do cross validation but at the same time we don't want to do the hyper parameter tunning in the train set because that's completly is not going to work well on out of sample data. The train set is completly observable to the model.

Doing cross validation in these 3 models the goal are to figure out the hyperparameters and the only one we are considering is alpha.

<code>**In polynomial regression remember that hyperparameter was the the degree of polynomial.**</code>


Because the hyperparameter in polynomial was discrete it was 1,2,3,4,5 etc. we could literally put in a for loop and plot the rmse of the test set and use the `elbow method` and figure out where is the optimal value for hyperparameter. However for alpha this is not discrete. And as we can see alpha can any number it can go from this case (although we defined 100 of them) but we know that if we want to go between 0 and 10 000 there can be millions of those alphas. Because it continues random variable. And for that reason the `elbow method` that could be used in polynomial regression cross validation maybe might not be best soliution. <code>**The computation is heavy!**</code>. But good news is that there is built-in function in python that is exploited in ridge/lasso/elastic net cross validation. Behind the scene that function is doing gradient descent to optimize this continuous hyperparameter.

`elbow method`

```python
from sklearn.model_selection import cross_val_score
```

```python
RMSE_CV = []
iterator = np.arange(0.0, 0.02, 0.001)
for i in iterator:
    MSE = - cross_val_score(
        estimator=ElasticNet(alpha=i), 
        X=X_train, 
        y=y_train, 
        cv=5, 
        scoring='neg_mean_squared_error'
    )
    RMSE_CV.append(np.sqrt(MSE).mean())
```

```python
output = pd.DataFrame(list(iterator), columns=['lambda_elasticnet'])
output['RMSE_CV'] = RMSE_CV
output.head()
```

We can do the `elbow method`

```python
def filter_correlations(df, bound):
    corr_matrix = df.corr().abs()
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
    tri_df = corr_matrix.mask(mask)
    to_drop = [c for c in tri_df.columns if any(tri_df[c] > bound)]
    reduced_df = df.drop(to_drop, axis=1)
    print("reduced dataframe has {} columns".format(reduced_df.shape[1]))
    return reduced_df
```
