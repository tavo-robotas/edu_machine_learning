## Plot a graph

```python
import matplotlib.pyplot as plt
x = [ i for i in range(10)]
y = [ i*2 for i in x]

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y);
```

```python
plt.scatter(x,y);
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(50, 70, 20)
y = np.linspace(160, 180, 20) + 15*np.random.normal(size=len(x))

plt.figure(figsize=(10,6))
plt.xlabel('Weight')
plt.ylabel('Height')
plt.scatter(x,y, s=80);
```
