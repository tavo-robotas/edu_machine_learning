## Plot a graph

```python
import matplotlib.pyplot as plt
x = [ i for i in range(10)]
y = [ i*2 for i in x]

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y);
```

```python
plt.scatter(x,y);
```
