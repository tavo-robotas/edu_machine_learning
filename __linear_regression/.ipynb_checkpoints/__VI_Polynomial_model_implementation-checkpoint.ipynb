Lets generate some data that is non linear

```python
x       = np.arange(-5, 5, 0.1) 
y       = 1 + 1*(x) + 1*(x**2) + 1*(x**3)
y_noise = 15 * np.random.normal(size=len(x))
y       = y + y_noise
plt.scatter(x, y)
```

```python
# x.ndim
# x.size
x.shape
```

```python
n      = len(x)
x_bias = np.ones((n, 1)) 
```

```python
x_bias.shape
```

```python
x = np.reshape(x, (n,1))
```

```python
x.shape
```

```python
X = np.append(x_bias, x , axis=1)
```

```python
X
```

\begin{multline*}
\theta = (X^{\top}X)^{-1} X^{\top}y
\end{multline*}

```python
X_T = np.transpose(X)
```

```python
X_prod = X_T.dot(X)
```

```python
X_INV = np.linalg.inv(X_prod)
X_PRE = X_T.dot(y)
```

```python
#from sklearn.linear_model import LinearRegression
```

```python
Î¸ = X_INV.dot(X_PRE)
Î¸

```

```python
plt.figure(figsize=(11,11))
plt.scatter(x, y)

# h(x) = Î¸_0 + Î¸_1 * x
# y = b + mx
plt.plot(x, Î¸[0] + Î¸[1] * x)
```

```python
def predict(x, intercept, slope):
    return  intercept + (x*slope)
```

```python
prediction = predict(x, Î¸[0], Î¸[1])
prediction
```

```python
def get_error(y_pred,y):    
    error = (y_pred - y) ** 2
    return np.sum(error) / len(error)
```

```python
error = get_error(prediction, y)
print(error)
```

Error is high. Letâs check what happens if we use polynomial equation. <br>

```python
x1_new.shape
```

```python
x_bias = np.ones((n,1))
```

2) Form the complete x matrix :

```python
x_n = np.append(x_bias, x1_new , axis=1)
x_n = np.append(x_n,    x2_new , axis=1)
x_n = np.append(x_n,    x3_new , axis=1)
```

```python
x_n[:10, :]
```

\begin{multline*}
\theta = (X^{\top}X)^{-1} X^{\top}y
\end{multline*}

```python
X_T = x_n.T
```

```python
X_prod = X_T.dot(x_n)
```

```python
X_INV = np.linalg.inv(X_prod)
X_PRE = X_T.dot(y)
```

```python
Î¸ = X_INV.dot(X_PRE)
Î¸
```
