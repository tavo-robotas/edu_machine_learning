## Learning model based on linear regression and gradient descent

In previous lectures we went through gradient descent algorithm and linear regression model and square error cost function.
Lets put together gradient descent with our cost function and that will produce an algoritmh for linear regresion, fitting a straight line to our data.



<code>**Gradient descent:**</code><br>

`repeat until convergence (for j = 0 and j = 1){
`\begin{multline*}
\theta_{j}:= \theta_{j} - \alpha \dfrac{\partial}{\partial\theta_{j}} J(\theta_{0}, \theta_{1})
\end{multline*}
`}`

<code>**Linear regression model with square error function:**</code><br>

\begin{multline*}h_{\theta}(x) = \theta_{0} + \theta_{1}x \end{multline*} <br>
\begin{multline*}J(\theta_{0}, \theta_{1}) = \dfrac{1}{2m}\sum_{i=1}^{m}(h_{\theta}(x^{(i)} ) -y^{(i)})^2\end{multline*}

We aim to apply gradient descent to minimize our square error cost function. In order to apply gradient descent the key term piece of code is the derivative term:

\begin{multline*}
\dfrac{\partial}{\partial\theta_{j}} J(\theta_{0}, \theta_{1})
\end{multline*}

We need to figure out what is this partial derivative term and plug in the definition of the cost function J. And turns out to be:

\begin{multline*}
\dfrac{\partial}{\partial\theta_{j}} J(\theta_{0}, \theta_{1}) = \dfrac{\partial}{\partial \theta_{j}} \cdot   \dfrac{1}{2m}\sum_{i=1}^{m}(h_{\theta}(x^{(i)} ) -y^{(i)})^2\ = \dfrac{\partial}{\partial\theta_{j}} \cdot \dfrac{1}{m}\sum_{i=1}^{m}( \theta_{0} + \theta_{1} x^{(i)} -y^{(i)})^2
\end{multline*}

We took our definition of hypothesis and plugged it in there, now we need to figure out what is this partial derivates for two case when J equal 0 and J equal 1.

\begin{multline*}
\theta_{0} = 0: \dfrac{\partial}{\partial\theta_{0}} J(\theta_{0}, \theta_{1}) = \dfrac{1}{m}\sum_{i=1}^{m}(h_{\theta} (x^{(i)}) - y^{(i)})
\end{multline*} 

\begin{multline*}
\theta_{1} = 1: \dfrac{\partial}{\partial\theta_{1}} J(\theta_{0}, \theta_{1}) =  \dfrac{1}{m}\sum_{i=1}^{m}(h_{\theta} (x^{(i)})- y^{(i)}) \cdot x^{(i)}
\end{multline*}

Computing those partial derivatives, going from one equation to simplification requires  some <code>**multivariate calculus**</code>. If it happens that you're less familiar with calculus don't worry about it. Its totally fine to take these equations <code>**which are just a slope of cost function J**</code> that we worked out and use it in implementation level of gradient descent. Lets plug them in back to gradient descent algorithm:

`repeat unlti convergence{`
\begin{multline*}
\theta_{0} := \theta_{0} - \alpha \dfrac{1}{m}\sum_{i=1}^{m}(h_{\theta} (x^{(i)}) - y^{(i)})
\end{multline*} 

\begin{multline*}
\theta_{1} = \theta_{1} - \alpha \dfrac{1}{m}\sum_{i=1}^{m}(h_{\theta} (x^{(i)})- y^{(i)}) \cdot x^{(i)}
\end{multline*}
`}`

This is gradient descent for linear regresion which repeats until convergence, theta parameters get updated after each substraction from alpha times derivative term. <code>**Don't forget when implementing gradient descent that we must simultaneously update theta parameters**</code>

Lets look how doest it work. There was one issue with gradient descent that it can be susceptible to local optima (local minimum) that depends from where you initialize theta parameters. We can end up at different local optima. But cost function for linear regresion is always* going to be a <code>**[convex](https://en.wikipedia.org/wiki/Convex_function) (bow shaped function) that doesn't have any local optima except for the one global optima and gradient descent will always converge to the global optimum (if you wait long enough and if the learning rate is not too high).**</code>
