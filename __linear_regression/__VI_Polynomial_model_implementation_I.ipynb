<code>**What's a polynomial**</code>

\begin{multline*}
\theta_{0} + \theta_{1}x + \theta_{2}x^2 + \theta_{3}x^3 + \ ... \ + \theta_{n}x^{n}
\end{multline*}

Basically our data <code>**x**</code> is taken to increasingly high powers and multplied by our coefficients <code>**ùúÉ**</code> that are scalling the data.

\begin{multline*}
3 - 5x + \dfrac{5}{6}x^{2} + 12.4x^{3}
\end{multline*}

\begin{multline*}
2x + x^{3}
\end{multline*}

Fiting this kind of model with all these nonlinearities seems complex but notice that <code>**coefficients are all linear**</code> Thus this is a standar linear model. It just the data that are getting raised to higher powers

## Polynomial regression

For example if we have a housing price data that might look like this:



There are <code>**DIFFERENT MODEL**</code> that we might fit to this. That means that our hypothesis function doesn't need not be linear <code>**a straight line**</code> if that does not fit the data well. Thus we can change the curve of our hypothesis function by making it a <code>**quadratic, cubic or square root function**</code> or any other polynomial form.

<code>**Quadratic**</code>
\begin{multline*}
\theta_{0} + \theta_{1}x + \theta_{2}x^2
\end{multline*}
<br>

<code>**Cubic**</code>
\begin{multline*}
\theta_{0} + \theta_{1}x + \theta_{2}x^2 + \theta_{3}x^3
\end{multline*}

<code>**Square root**</code>
\begin{multline*}
\theta_{0} + \theta_{1}x + \theta_{2}\sqrt{x} + \theta_{3}x^3
\end{multline*}

```python
x       = np.arange(-5, 5, 0.45) 
y       = 1 + 1*(x) + 1*(x**2) + 1*(x**3)
y_noise = 30 * np.random.normal(size=len(x))
y       = y + y_noise
plt.figure(figsize=(10, 6))
plt.scatter(x, y)
```

```python
def first_order(x):
    return 1 + 2*x 

def second_order(x):
    return 1 + 2*x + 2*x**2

def third_oder(x):
    return 1 + 2*x + 2*x**2 + 1*x**3

def forth_order(x):
    return 1 + 2*x + 2*x**2 + 1*x**3 + 1*x**4
```

```python
y_1 = first_order(x)
y_2 = second_order(x)
y_3 = third_oder(x)
y_4 = forth_order(x)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10,8))
fig
ax[0, 0].plot(x, y_1)
ax[0, 0].scatter(x, y, c='r')
ax[0, 0].set_title("1'st order")
  
ax[0, 1].plot(x, y_2)
ax[0, 1].scatter(x, y, c='r')
ax[0, 1].set_title("2'nd order")
  
ax[1, 0].plot(x, y_3)
ax[1, 0].scatter(x, y, c='r')
ax[1, 0].set_title("3'rd order")
  
ax[1, 1].plot(x, y_4)
ax[1, 1].scatter(x, y, c='r')
ax[1, 1].set_title("4'th order")
plt.show()
```

## Which order should we use ?

<code>**Bayes information criterion**</code>

\begin{multline*}
BIC_{k} = n\log(SS_{\epsilon}) + k\log(n)
\end{multline*}

This is a quantity that we can compute for every model order, up to <code>**k**</code> model orders. 

- <code>**n**</code> is data points
- <code>**ùëÜùëÜùúñ**</code> is the sum of squares of residuals

\begin{multline*}
SS_{\epsilon} = \sum^{n}_{i=1}(\hat{y} - y)^{2}
\end{multline*}
