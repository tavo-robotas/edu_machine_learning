# Regularization

## 
We went so far and we really made our models complex, to complex in certain cases. And this led us to an issue where complexiti made the model overfit the training data. Now we need to constrain model features that overfitt data and makes itself a little simple on its own. In other words we will penalize higher degree of freedom weights who are making it overfit the data.

Consider a model with only one featue <code>**w_1**</code> and bias <code>**b**</code>. If we limit b to be 0, then the model will be little simpler as it only depends on <code>**w_1**</code>. But what if we limit both <code>**w_1**</code> and <code>**b**</code> to be small. The model will be complex when compared to model wit <code>**b=0**</code> but simple compared with no constrains on <code>**w_1**</code> and <code>**b**</code>.

Even though there are many types of regularization out tere, we will be looking int two main types:

## L1 ( Lasso ) - Least Absolute Shrinkage and Selection Operator
## L2 ( Ridge ) - "Squared magnitude"
## Elastic net  - Combination of both

And the difference between these two L1 and L2 is the <code>**penalty term**</code>.

## L1 ( Lasso )

L1 regularization represents the addition of a regularized term – few parts of l1 norm, to the loss function (mse in case of linear regression). Here, if weights are represented as w0, w1, w2 and so on, where w0 represents bias term, then their l1 norm is given as:

<code>**Lasso**</code>:
\begin{multline*}
\text{L1 norm} \ = \sum_{i=1}^{m}|w_i|
\end{multline*}


We add a few parts of this regularized term to our loss function. Hence, we multiply a regularization coefficient α to this term before adding it to the loss function, making our final loss function:

\begin{multline*}
J(w; x, y) = MSE(w; x, y) + \alpha \sum_{i=1}^{m}|w_i|
\end{multline*}

<code>**If α is close to 0**</code>, this essentially becomes a Linear Regression model while  <code>**if α is too large**</code>, then every w becomes 0 giving us a flat line.

<code>**Regularization gives us a sparse model!**</code> And what does that mean ?

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Lasso
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error

m = 100
x = 5 * np.random.rand(m, 1) - 2
y = 0.7 * x ** 2 - 2 * x + 3 + np.random.randn(m, 1)

plt.scatter(x, y)
plt.show()
```

```python
def get_preds_lasso(x1, x2, alpha):
    if alpha == 0:
        # Some issues with alpha = 0 in Lasso.
        # We know that alpha=0 is essentially LinearRegression
        # so we will use it
        model = Pipeline([
            ('poly_feats', PolynomialFeatures(degree=16)),
            ('lasso', LinearRegression())
        ])
    else:
        model = Pipeline([
            ('poly_feats', PolynomialFeatures(degree=16)),
            ('lasso', Lasso(alpha=alpha))
        ])
    model.fit(x1, x2)
    return model.predict(x1)

alphas = [0, 0.7, 5]
cs = ['r', 'g', 'b']

plt.figure(figsize=(10, 6))
plt.plot(x1, x2, 'b+', label='Datapoints')

for alpha, c in zip(alphas, cs):
    preds = get_preds_lasso(x1, x2, alpha)
    # Plot
    plt.plot(sorted(x1[:, 0]), preds[np.argsort(x1[:, 0])], c, label='Alpha: {}'.format(alpha))

plt.legend()
plt.show()
```
