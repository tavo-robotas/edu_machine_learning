## Cost function

<code>**Logistic regression**</code>

\begin{multline*}
J(\theta) = -\dfrac{1}{m} [ \sum_{y=1}^m y^{(i)}log(h_{\theta}(x^{(i)})) + (1-y^{(i)})log(1 - h_{\theta}(x^{(i)}))] + \dfrac{\lambda}{2m} \sum_{j=1}^n \theta_{j}^2
\end{multline*}

The cost function that we will use for NN is <code>**going to be a generalization of the one we used for logistic regression**</code>. Where instead of having one output unit for 'th element we may have  of them. 

<code>**Neural network**</code>

\begin{multline*}
h_{\Theta}(x) \in \mathbb{R}^{K}, \ (h_{\Theta}(x))_{i} = \ i^{\text{th}} \ \text{output}
\end{multline*}

\begin{multline*}
J(\Theta) = -\dfrac{1}{m} [ \sum_{y=1}^m \ \sum_{k=1}^K y_{k}^{(i)}log(h_{\theta}(x^{(t)}))_{k} + (1-y_{k}^{(t)})log(1 - h_{\theta}(x^{(t)})_{k})] + \dfrac{\lambda}{2m} \sum_{l=1}^{L-1} \ \sum_{i=1}^{s_{l}} \ \sum_{j=1}^{s_{l+1}}\ (\Theta_{j, i}^{(l)})^2
\end{multline*}

<code>**Note:**</code>

 * double sum simply adds up the logistic regression costs calculated for each cell in the output layer
 * triple sum simply adds up the squares of all the individual s in the entire network.
 * i in the triple sum does not refer to training example i

Our neural network now outputs vectors in  where  migh be equal to 1 if we have a binarry classification problem. This notation  (()) is used to denothe the 'th output. That is () is a K dimensional vector and so this subscript  just selects out the 'th element of the vector that is output by our neural network. 

() our cost function is a similar term to what we have for logistic regression, except that we have the sum from k=1 through k elements. This summation is basically a sum over our  output units.<br>

\begin{multline*}
\sum_{k=1}^K
\end{multline*}

Let say we have 4 output units that is in the final layer of our NN so this is going to be a sum from k=1 through 4 of the logistic regression algorithm's cost function but summing that cost function over each of our 4 output units in turn. 

Notice that in particular that this applies to (), ((())) because basically we are taking the K output unit and comparing that to the value of  which is that one of those vectors saying what cost it should be.

\begin{multline*}
+ \dfrac{\lambda}{2m} \sum_{l=1}^{L-1} \ \sum_{i=1}^{s_{l}} \ \sum_{j=1}^{s_{l+1}}\ (\Theta_{ji}^{((l)})^2
\end{multline*}

And finally the second <code>**regularization term**</code> similar to what we had for the logistic regression. This summation term although looks really complicated but all its doing is summing over these terms
\begin{multline*}
\Theta_{j,i}^{(l)}
\end{multline*}

for all values of , , , except that we dont sum over the terms corresponding to the <code>**bias values**</code> like we had for logistic regression
\begin{multline*}
\theta_{0}
\end{multline*}

<code>**So we don't sum over the terms to where  is equal to 0**</code>. So that is because when we are computing the activation of a neuron we have terms like these.
\begin{multline*}
\Theta^{(j)}_{i0}x_{0} + \Theta^{(j)}_{i1}x_{1} + \ ... 
\end{multline*}


Values with a zero there that correspond to something that multiplies into x0 or a0 this is kinda like a bias unit and by analogy to what we were doing for logitic regression we won't sum over those term in our regularization term, because we don't want to regularize them and strip their values to zero.

This is one possible convention and even if we were to sum over  = 0 up to  it would work  about the same and doesn't make a big difference. But maybe this convention of not regularizing the bias term is more common.

Implementing either advanced framework or custom made learning algorithm ussualy the routine with cost function is:
 * 1) we provide it with some <code>**matToVec**</code> product  thats has been unrolled into a vector
 * 2) reshape to pull out the elements from vector to get back our original parameters
 * 3) now we have more conveniant form t run forward propagation and back propagation to compute our derivates D1, D2 and J()
 * 4) finally then we can take our derivatives D1, D2 and unroll them keeping the same ordering as we did when unrolled our 's to get gradient vector of these derivatives

\begin{multline*}
J(\Theta) = -\dfrac{1}{m} \sum_{y=1}^m \ \sum_{k=1}^K[y_{k}^{(i)}log(h_{\theta}(x^{(i)}))_{k} + (1-y_{k}^{(i)})log(1 - h_{\theta}(x^{(i)})_{k})] + \dfrac{\lambda}{2m}[ \sum_{j=1}^{25} \ \sum_{k=1}^{400}\ (\Theta_{j, k}^{(1)})^2 \ \sum_{j=1}^{10}\sum_{k=1}^{25}\ (\Theta_{j, k}^{(2)})^2]
\end{multline*}

```python
def cost(nn_params, input_layer_size, hidden_layer_size, num_labels, X, y, =1):
    
    m = len(y)
    _slice = hidden_layer_size * (input_layer_size + 1)
  
    th1 = np.reshape(
        nn_params[: _slice], (hidden_layer_size, input_layer_size  + 1), order = 'F')

    th2 = np.reshape(
        nn_params[ _slice:], (num_labels,        hidden_layer_size + 1), order = 'F')
    
    h = forwardprop(th1, th2, m, X)
    
    # cost 
    cost = np.sum( (np.multiply(y, np.log(h))) + (np.multiply(1-y, np.log(1-h))) )
    
    # regularization
    
    if  != 0:
        L2_on_TH1 = np.power(th1[:,1:],2)
        L2_on_TH2 = np.power(th2[:,1:],2)

        reg_sum_1 = np.sum(np.sum(L2_on_TH1, axis = 1))
        reg_sum_2 = np.sum(np.sum(L2_on_TH2, axis = 1))
        reg_term  = (reg_sum_1 + reg_sum_2) * lmbda / (2*m)
    
    J = np.sum(cost) / -m  + reg_term
    
    return J
```

```python
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))
```

<code>**Forward propagation**</code><br>

\begin{multline*}
a^{(1)} = x
\end{multline*}
\begin{multline*}
z^{(2)} = \Theta^{(1)}a^{(1)}
\end{multline*}
\begin{multline*}
a^{(2)} = g(z^{(2)}) \ (\text{add} \ a_0^{(2)})
\end{multline*}
\begin{multline*}
z^{(3)} = \Theta^{(2)}a^{(2)}
\end{multline*}
\begin{multline*}
a^{(3)} =  g(z^{(3)}) = h_{\Theta}(x)
\end{multline*}

```python
def forwardprop(theta1, theta2, m, X):
     bias = np.ones((m,1))
     a1   = np.hstack((bias, X))
     z1   = a1.dot(theta1.T)
     a2   = sigmoid(z1)
     a2   = np.hstack((bias, a2))
     z2   = a2.dot(theta2.T)
     a3   = sigmoid(z2)
     h    = a3;
     return h
```
