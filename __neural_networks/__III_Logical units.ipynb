## Non linear classification example : XOR/XNOR

Consider the following problem: We have input features
<code>**𝑥1, 𝑥2 are binary (0,1)**</code>



Let say it is a simplified version of a more complex learning problem where we might have bunch of positive examples and negative examples. And we want to learn non linear division boundary that seperates our samples



Lets rather use the simpler to examine example for now.</code>



This is really computing the type of label y equals to 𝑥1 <code>**XOR**</code> 𝑥2 <br> or the 𝑥1 <code>**XNOR**</code> 𝑥2 function. <code>**XNOR**</code> is alternative notation for <code>**NOT**</code>  𝑥1 <code>**XOR**</code> 𝑥2 

\begin{multline*}
y =
\end{multline*}

\begin{multline*}
\ \ \ \ \ \ \ \ x_{1} \ \text{XOR} \ x_{2}
\end{multline*}

\begin{multline*}
\ \ \ \ \ \ \ \ x_{1} \ \text{XNOR} \ x_{2}: \ \text{NOT} \ (x_{1} \ \text{XOR} \ x_{2})
\end{multline*}


It turns out that these specific sample in the use works out a little bit if we use <code>**XNOR**</code> function instead.<br>
So we going to have positive examples of either <code>**both are true or both are false**</code>

<code>**y equals 0 if only one of them is true**</code>, so we want to figure out if we can get a neural newtork to fit to this sort of training set. In order to build uo to a network that fits the <code>**XNOR**</code> example we are going to start with slighly simple task and try to fit the <code>**AND**</code> function.

## AND

Lets say we have inputs  <code>**𝑥1,𝑥2**</code> that are binaries and our target  <code>**label 𝑦**</code> is <code>**𝑦 = 𝑥1 AND 𝑥2**</code>. This is logical AND


\begin{multline*}
x_{1}, x_{2} \in {0, 1}
\end{multline*}

\begin{multline*}
y = x_{1} \text{AND} \ x_{2} 
\end{multline*}



To get a one unit network to compute this <code>**logical AND function**</code> lets assign some vales to the weights (parameters) of this network. This means we are assigning a value of -30 to the valua associated with 𝑥0 and the same for 𝑥1, 𝑥2.

This way we are saying that our hypothesis

\begin{multline*}
h_{\Theta}(x) = g(-30 + 20x_{1} + 20x_{2})
\end{multline*}

\begin{multline*}
\Theta =
\begin{bmatrix}
\theta_{10}^{(1)} \\
\theta_{11}^{(1)}\\ 
\theta_{12}^{(1)}
\end{bmatrix}
=
\begin{bmatrix}
-30 \\
20 \\ 
20
\end{bmatrix}
\end{multline*}

Let's lok at what this little single neuron network will compute. Let look at the four possible input values for 𝑥1, 𝑥2 and look at what the hypotheses output will be in that case.<code>**Lets write out truth table.**</code>

* <code>**x1=0  and  x2=0  then  g(−30)≈0**</code><br>
* <code>**x1=0  and  x2=1  then  g(−10)≈0**</code><br>
* <code>**x1=1  and  x2=0  then  g(−10)≈0**</code><br>
* <code>**x1=1  and  x2=1  then  g(10) ≈1**</code><br>

So this is computing <code>**ℎΘ(𝑥) ≈ 𝑥1 AND 𝑥2**</code>. In other words it <code>**outputs 1 if and only if 𝑥1 AND 𝑥2 are both equal to 1**</code>.

## OR

This network computes the <code>**logical OR function.**</code>

This way we are saying that our hypothesis

\begin{multline*}
h_{\Theta}(x) = g(-10 + 20x_{1} + 20x_{2})
\end{multline*}

Let continue or our complex non linear hypotheses. We went through <code>**𝑥1 AND 𝑥2**</code>, <code>**𝑥1 OR 𝑥2**</code> when 𝑥1,𝑥2 are binary. But we also can have a network to compute <code>**negation**</code>, that is to compute the function <code>**NOT 𝑥1**</code><br>
So we have only one input feature (and +1 bias unit), lets add weights.

## NOT

\begin{multline*}

<code>**(NOT 𝑥1) AND (NOT 𝑥2)**</code> if we want to compute a function like this part of that will probably be putting large negative weights in fron of 𝑥1 and 𝑥2, but it should be feasible to get neural network with just one output unit to compute this as well. ## XNOR

Now lets put all three pieces together that we have put together as the network sfor computing <code>**AND, NOT, OR**</code> we should be able to put all these three together to compute <code>**XNOR**</code> function.



<code>**Thus ℎΘ(𝑥) is equal to 1 when  either both 𝑥1 = 𝑥2 = 0 or 𝑥1 = 𝑥2 = 1**</code>. We end up with nonlinear decision boundary that computes this <code>**XNOR**</code> function. This primary intuition why neural networks can compute pretty complicated functions. When we have multiple layers intially we have relatively simple functions of the inputs and the second layer but the third layer can build on that to compute even more complex functions and the layer after that even more.

## Coding a logical units (neurons)

```python
import numpy as np

class Neuron:
    
    def __init__(self, weights, bias):
        self.__weights = weights
        self.__bias    = bias
    
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    
    def propagate(self, X):
        z = np.dot(self.__weights, X) + self.__bias
        return self.sigmoid(z)
    
```

<code>**That’s it!**</code>. Lets make network

```python
class NN:
    '''
        A neural network with:
            - 2 inputs x1, x2 + bias input
            - hidden layer with 2 units (a2_1, a2_2)
            - output layer with 1 unit  (a3_1)
    '''

    def __init__(self, weights, bias):
        self.__weights = weights
        self.__bias    = bias
        
        self.a2_1 = Neuron(self.__weights[0], self.__bias)
        self.a2_2 = Neuron(self.__weights[1], self.__bias)
        self.a3_1 = Neuron(self.__weights[2], self.__bias)
        
    def forward(self, X):
        out_a2_1 = self.a2_1.propagate(X)
        out_a2_2 = self.a2_2.propagate(X)
        out_a3_1 = self.a3_1.propagate(np.array([self.__bias, out_a2_1, out_a2_2]))
        
        return out_a3_1
```

```python
def resize(image, width = None, height = None, inter = cv.INTER_AREA):
    dim = None
    (h, w) = image.shape
    if width is None and height is None:
        return image
    if width is None:
        r = height / float(h)
        dim = (int(w * r), height)
    else:
        r = width / float(w)
        dim = (width, int(h * r))
    resized = cv.resize(image, dim, interpolation = inter)
    return resized
```
