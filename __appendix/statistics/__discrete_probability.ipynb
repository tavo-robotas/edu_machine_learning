# Discrete probability

Let's start off with a definition for discrete uniform distributions.<br>

<code>**Definitions:**</code><br>
<code>**S - is a set called the sample space**</code>

\begin{multline*}
E \subseteq S
\end{multline*}

<code>**E is a subset called an event**</code><br>

The probability of an event E occuring is:

\begin{multline*}
P(E) = \dfrac{|E|}{|S|} 
\end{multline*}

Its the number of elements in E divided by the number of elements in S.

Simple problem if we toss two fair coins. What is the probability of getting <code>**exactly one head**</code>?

<code>**S = { HH, HT, TH, TT }**</code><br>
<code>**E = { HT, TH}**</code><br>

\begin{multline*}
P(E) = \dfrac{|E|}{|S|} : \dfrac{2}{4} = \dfrac{1}{2}
\end{multline*}

\begin{multline*}
\text{Given} \ P(A) \ \text{, we can find} \ P(\overline{A})
\end{multline*}

Meaning the probability of <code>**NOT A**</code> or <code>**A complement**</code> - that is everything that is not in that set.

\begin{multline*}
P(A) = 1 - P(\overline{A})
\end{multline*}

What us the probability of not getting two head<br>

\begin{multline*}
P(HH) = \dfrac{|{HH}|}{|S|} : \dfrac{1}{4}
\end{multline*}


\begin{multline*}
P(\overline{HH}) = 1 - \dfrac{1}{4} = \dfrac{3}{4}
\end{multline*}

## Combinatorics problem

Given a deck of <code>**52 cards**</code>, we draw <code>**5**</code>. What is the prob. of getting:
 - A) 3 Aces, 2 Jacks ?
 - B) 3 Aces and a pair ?

Event of 3 Aces and 2 Jacks: <br>

\begin{multline*}
E = (\dfrac{4}{3}) \ (\dfrac{4}{2})
\end{multline*}

Sample space is the all posible combinations of cards we can draw.

\begin{multline*}
S = (\dfrac{52}{5})
\end{multline*}

Probability of A:

\begin{multline*}
P(A) = \dfrac{(\dfrac{4}{3})(\dfrac{4}{2})}{\dfrac{52}{5}}
\end{multline*}

Probability of B:

\begin{multline*}
P(A) = \dfrac{(\dfrac{4}{3})(\dfrac{12}{1})(\dfrac{4}{2})}{\dfrac{52}{5}}
\end{multline*}

12/1 because there is 13 sets, but we can't pick aces so we exclude them from entire set and we pick 1. Then 4/2 is that we have to choose 2 (pair) of 4 cards.

## Axioms of probability

Let <code>**S**</code> be the sample space and <code>**Ai , Aj**</code> be event in <code>**S**</code> there can be infinite amount of them.

<code>**1)**</code> 
\begin{multline*}
P(A) \geq 0
\end{multline*}

<code>**2)**</code>
\begin{multline*}
P(S) = 1
\end{multline*}

<code>**3)**</code> 
\begin{multline*}
\bigcap A_{i} = \emptyset \ \text{then} \bigcup P(A_{i}) = P(A_{1}) + P(A_{2}) + \ ...
\end{multline*}

Intersection of all event is equal to empty set. Then the union of the probability of all of the events is just equal to sumed probabilities.<br>

\begin{multline*}
A \cap B = \emptyset \ \text{then} \ P(A\cup B) = P(A) + P(B)
\end{multline*}

We have A and B there are no overlaps therefor if we want A or B there is just the probability of P(A) and plus the probability of P(B). There are no overlaps so we don't have to subtract anything.

So by using these three axioms we will prove the <code>**compliment rule**</code>

\begin{multline*}
P(\overline{A}) = 1 - P(A)
\end{multline*}

Going throught this step by step. First we want a sample space
\begin{multline*}
S = A \cup \overline{A}
\end{multline*}

When we take intersection

\begin{multline*}
A \cap \overline{A} = \emptyset
\end{multline*}

We get <code>**empty set**</code> because there is <code>**not intersection**</code>

So by axiom <code>**3**</code> we kow that because A interesetion of not A is the empty set the probability of A or not A is equal to probability of A plus the probability of not A. 

\begin{multline*}
P(A \cup \overline{A}) = P(A) + P(\overline{A})
\end{multline*}

And now from this the probability of A or not A is actually <code>**equal to the probability of the sample space**</code>. 

\begin{multline*}
P(S) = P(A) + P(\overline{A})
\end{multline*}

<code>**Axiom 2**</code> says that the probability of the sample space is always equal to one

\begin{multline*}
P(S) = 1
\end{multline*}

Then by using simple algebra we can state

\begin{multline*}
1 = P(A) + P(\overline{A}) 
\end{multline*}

\begin{multline*}
P(\overline{A}) =  1 - P(A)
\end{multline*}



## The aditiive rule

What it says here is that the probability of A or B is equal to probability of A intersection not B plus the probability of B
\begin{multline*}
P(A\cup B) = P(A\cap \overline{B}) + P(B)
\end{multline*}

## Inclusion and exlusion principle
