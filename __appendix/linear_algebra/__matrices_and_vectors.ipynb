## Practical example in machine learning

I call it <code>**ONE SHOOT**</code> method.
This is very efficient and simplified way to deal with machine learning problems, we will get more into to this approach on <code>**vectorization**</code> topics.
Lets take a look into a neat mathematical trick. We have samples of some house sizes and a  hypothesis that predicts the house price. And we want to compute h(x) for each of the house. We can do this in computationally efficient way by using <code>**matrix vector multiplication**</code>

<code>**sizes**</code>:
<code>50</code>
<code>77</code>
<code>52</code>
<code>218</code>


$$ h_{\theta}(x) = -0.1 + 5x$$

Lets construct a matrix of our data samples and a vector of hypothesis parameters (coefficients) and multiply them together.

\begin{align}
\begin{bmatrix}
1 & 50 \\
1 & 77 \\
1 & 52 \\
1 & 218
\end{bmatrix}
\times
\begin{bmatrix}
-0.1 \\
5
\end{bmatrix}
=
\end{align}


When implementing this in software when you some data samples and we want to use our hypothesis to predict the prices, the value of y of all of these data sample entries.  We can do this in just few lines of code we only need a appropriate library to do linear algebra operations like matrix multiplication. As an alternative we can do this with loops and other bunch of work which can be quite cumbersome. By using linear algebra in any various programming languages it allows us not only to simply the code but for subtle reasons it turns out to be much more computationally efficient to make predictions on all data samples.

```python
import numpy as np

D = np.array([
    [1, 50],
    [1, 77],
    [1, 52],
    [1, 218]
])

θ = np.array([-0.1, 5])

##############
print("D matrix properties")
print(type(D))
print(D.ndim)
print(D.shape)
##############
print("θ vector properties")
print(type(θ))
print(θ.ndim)
print(θ.shape)
```

```python
## in math it's called dot product of matrix and vector or another matrix
y = D.dot(θ)
y
```

## Matrix multiplication properties

Matrix multiplication is highly usefull since we can pack a lot of computation into just single data structure operation. There are few properties of this operation that we should be aware and carefull when using this method. 

### Commutative property (komutacinis)

<code>**Matrix multiplication IS NOT commutative**</code><br>
When working with real numbers , scalars, multiplication is commutative and the order doesn't matter.

\begin{multline*}
3 \times 2 = 2 \times 3
\end{multline*}

<code>**But this is not true for matrix multiplication**</code><br>
Let A and B be matrices

\begin{multline*}
A \times B \neq B \times A
\end{multline*}

It's not good to arbitrarily reverse the order in which you multiply matrices.<br>
If we swap order around of these two matrices we get different soliutions.

\begin{multline*}
\begin{bmatrix}
1 & 1 \\
0 & 0
\end{bmatrix}
\times
\begin{bmatrix}
0 & 0 \\
2 & 0
\end{bmatrix}
=
\begin{bmatrix}
2 & 0 \\
0 & 0
\end{bmatrix}
\end{multline*}


\begin{multline*}
\begin{bmatrix}
0 & 0 \\
2 & 0
\end{bmatrix}
\times
\begin{bmatrix}
1 & 1 \\
0 & 0
\end{bmatrix}
=
\begin{bmatrix}
0 & 0 \\
2 & 2
\end{bmatrix}
\end{multline*}

In general reversing the order of matrix multiplication can even change the dimensions of outcome.
If <code>**A x B**</code> is <code>**mxn and nxm**</code>, then <code>**A x B**</code> is going to be <code>**mxm**</code> dimensions structure matrix and <code>**B x A**</code> is going to be <code>**nxn**</code> dimensions structure. 

### Associative property (asociatyvus)

When dealing with scalars (real number) we have associative properties pf real numbers multiplication.

\begin{multline*}
3 \times 2 \times 2 = (3 \times 2) \times 2 = 3 \times (2 \times 2) = 12
\end{multline*}

<code>**Matrix multiplication iS associative**</code><br>
\begin{multline*}
A \times B \times C = A \times (B \times C) = (A \times B) \times C
\end{multline*}

\begin{multline*}
D = B \times C, A \times D
\end{multline*}


\begin{multline*}
E = A \times B, E \times C
\end{multline*}

Both of these option will give us the same soliution.

## Identity matrix

When dealing with scalars we can think of this as identity of multiplication. 
Lets say we multiply any real number (a) by 1. 

\begin{multline*}
1 \times a = a \times 1 = a
\end{multline*}

This is identity operation and it has to satisfy this equation. 
In linear algebra regarding matrices there is <code>**identity matrix**</code> denoted <code>**I**</code> and it's a <code>**square matrix**</code> meaning its dimensions are <code>**nxn**</code>.

```python
import numpy as np

A = np.eye(8)
A
```

It has a property that it has <code>**1's along the diagonals (lt. istrižainės) and 0's everywhere else**</code>.

For any matrix A

\begin{multline*}
A \times I = I \times A = A
\end{multline*}

## Inverse and transpose

How does inverse relate to real numbers. Remember that 1 number plays a role of <code>**identity**</code> in the space of real numbers because well its obvious that multiply by anything is equal to itself. And there is a property that each real number has an <code>**inverse**</code> for example 3·(3ⁿ) = 1 , where n=-1. (3ⁿ) where n=-1 is the same as 1/3. It can be any real number the product is always going to be the identity.<br>
<code>**But in the space of real number not everything has an inverse**</code> like 0, its inverse 1/0 would be undefined.
So how does this apply to matrix and how to computer inverse of matrix. The idea is this:

\begin{multline*}
AA^{-1}= A^{-1}A = I
\end{multline*}

If A is an <code>**mxm matrix a.k.a square matrix**</code> and if it has an inverse then its product is equal to identity matrix.

\begin{multline*}
\begin{bmatrix}
3 & 4 \\
2 & 16
\end{bmatrix}^{-1}
 = 
\dfrac{1}{3 \times 16 - 4 \times 2 }
\times
\begin{bmatrix}
16 & -4 \\
-2 & 3
\end{bmatrix}  
= 0.025 \times
\begin{bmatrix}
16 & -4 \\
-2 & 3
\end{bmatrix}  
=
\begin{bmatrix}
0.4 & -0.1 \\
-0.05 & 0.075
\end{bmatrix}  
\end{multline*}

But these days rarely who does it by hand, most of software programming languages linear algebra libraries has built-in tools to handle this operation.

```python
import numpy as np

A  = np.array([
    [3, 4 ],
    [2, 16]
])
A
```

```python
B = np.linalg.inv(A)
B
```

<code>**Matrices that don't have an inverse are "singular" or "degenerate"**</code>

Other special matrix operation we will go though is <code>**transpose**</code>

\begin{multline*}
A = 
\begin{bmatrix}
1 & 2 & 0 \\
3 & 5 & 9
\end{bmatrix}, \
B =  A^{\top} =
\begin{bmatrix}
1 & 3 \\
2 & 5 \\
0 & 9
\end{bmatrix}
\end{multline*}

<code>**Just convert rows into columns**</code> Another way to think about transpose is to image a 45° axis from left to right that crosses matrix and we are mirroring or flipping along that axis our matrix values.

\begin{multline*}
\text{Let A be an matrix m} \times \text{n and let B} = A^{\top}. 
\text{Then B is an n} \times \text{m matrix and} \ B_{ij} = A_{ji}
\end{multline*}

\begin{multline*}
B_{12} = A_{21} = 3
\end{multline*}

```python
A  = np.array([
    [1, 2, 0],
    [3, 5, 9]
])

A
```

```python
B = A.T
B
```

```python
B = np.transpose(A)
B
```
