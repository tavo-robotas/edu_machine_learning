KNN - K Nearest Neighbors is one of the simplest <code>**supervised**</code> machine learning algorith mostly used for <code>**classification problems**</code>. It classifies a data point base on how is neigbors are classified. It stores all available clases and classifies new cases based on a <code>**similarity measure**</code>.<br>

<code>**k**</code> in KNN is a parameter that refers to the number of nearest neighbors to include in the majority voting process. 

If we estimate new sample we want to know what the neighbors are. If <code>**k=3**</code> a data point in question is <code>**classified by majority votes from its 3 nearest neighbors**</code>. So this umknowed sample would be classified as <code>**o**</code> since our three of three neighbors are o. 

Why previously our <code>**k**</code> was 3 , not 11 or 50 ? Since KNN algorithm is based on feature similarity chossing the right value of k is a process called <code>**parameter tuninig**</code> and its important for better accuracy

Trying to evaluate if sample in question is a square or a triangle and k equals three. We will look ant three nearest neighbors and we will say <code>**its a square**</code>. If we put <code>**k=7**</code> we classify as a  <code>**triangle**</code> depending on what the other data is around and you can see if the k changes depending on where that point is that drastically changes our answer.

If k is to low the bias is to noisy it's right next to a couple of samples and it's gonna pick those things. But if its to big it will take a lot of time to proccess.<br>
<code>**To choose a value of factor k:**</code>
 * <code>**square root of n**</code>, where n is the total number of samples
 * <code>**odd values of k**</code> is selected to avoid confusion between two classes of data, otherwise we will have balance between two different factors that are equal.

On the basis of the given data we have to classify <code>**data point(57kg, 170cm)**</code> as <code>**normal or underweight**</code> using KNN. To find the nearest neighbors we will calculate <code>**[Euclidean distance](https://en.wikipedia.org/wiki/Euclidean_distance#:~:text=In%20mathematics%2C%20the%20Euclidean%20distance,being%20called%20the%20Pythagorean%20distance.)**</code><br>

\begin{multline*}
d(a,b) = \sqrt{\sum_{i=1}^n(b_{i} -a_{i})^2}
\end{multline*}


\begin{multline*}
p1 = (1, 1), \ p2 = (5, 4) \
ed = \sqrt{(5-1)^2 + (4-1)^2} = 5
\end{multline*}

There is as well another <code>**Manhattan distance**</code> method

\begin{multline*}
md = |5-1| + |4-1| = 7
\end{multline*}

<code>**Euclidean and Manhattan distance main difference is that Euclidean distance is nothig but a direct and least possible distance between two points, whereas Manhattan distance is a distance between points measured along the axis at right angle**</code>



Lets calculate it to understand clearly. We have some unknowned point and we have our other data points where the data is scattered around. Calculate distance d1, d2, ..., d11.

* A positive integer k is specified, along with a new sample
* We select the k entries in our database which are closest to the new sample
* We find the most common classification of these entries
* This is the classification we give to the new sample

## Predict diabetes

Our objective to predict whether a person will be diagnosed with diabetes or not. We have a dataset that you can [download here](https://www.kaggle.com/saurabh00007/diabetescsv) from <code>**kaggle datasets**</code> of <code>**768 people**</code> who were or were not diagnosed with diabetes.

```python
import math
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score
```

```python
df = pd.read_csv('diabetes.csv')
df
```

Looking at <code>**Insulin**</code> we can conclude thats why these people are taking it, because have diabetes. And that could cause issues in some of the machine learning models but for our basic setup this will work fine for KNN. 

```python
df.info()
```

Detect missing values, none, NaN 

```python
df.isna().sum()
```

Results of distinct observations in our dataframe columns

```python
df.nunique()
```

Values of columns like <code>**Glucose, BloodPressure**</code> can not be accepted <code>**as 0**</code> because it will affect the outcome, we will replace such values with the mean of the respective column, in other words replace each instance of 0 in some of the feature columns with the average person data. We can do it with pandas 
<code>**[replace](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html)**</code> method.  This trick is a way that we can preserve our samples and we can do our computations this just brings those missing values out of the equation.

```python
no_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'BMI', 'Insulin']

for col in no_zero:
    df[col] = df[col].replace(0, int(df[col].mean(skipna=True)))

df
```

Before proceeding further we have to splot our dataset into train and test

```python
X = df.iloc[:, 0:8]
y = df.iloc[:,   8]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2, stratify=y)
```

<code>**Recap that scaling is important for any algorithm that computes distance or asummes normality, we have to scale our features**</code>

```python
scaler  = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test) 
```

<code>**fit_transform() and transform()**</code>. 



Then we define the model using <code>**[KNeighborsClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html)**</code> and fit the train data in the model with our parameters:
 * <code>**n_neighbors is k value**</code>
 * <code>**p is the power parameter to define the metric used, p = 1 is manhattan and p = 2 is euclidean**</code>
 * <code>**weights, default 'uniform'**</code> means all points in each neighborhood are weighted equally. 'distance' weight points by the inverse of their distance. In this case, closer neighbors of a query point will have a greater influence than neighbors which are further away.

```python
clf = KNeighborsClassifier(n_neighbors=11, p=2, metric='euclidean')
clf.fit(X_train, y_train)
```

To define n_neighbors or k value this what we could do

```python
k = math.sqrt(len(y_test))
k = math.floor(k) - 1
```

```python
y_pred = clf.predict(X_test)
y_pred
```

For classification model evaluation we will use <code>**confusion matrix**</code>

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
def plot_conf_matrix(matrix, title="land of confusion", cmap='cool'):
    plt.imshow(matrix, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(1)
    plt.xticks(tick_marks, ["0","1"])
    plt.yticks(tick_marks, ["0","1"])
    plt.tight_layout()
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    width, height = matrix.shape
    
    for x in range(width):
        for y in range(height):
            plt.annotate(str(matrix[x][y]), xy=(x, y), horizontalalignment='center', verticalalignment='center')
```

```python
plot_conf_matrix(cm)
```

f1 score takes into both side of the balance of false positives where accuracy looks at just how many we got right out of how many we got wrong. Accuracy is always better in numbers than f1 score, but the f1 is more telling that let's us know that there is more false positives than we would like on here

```python
print(f'f1 score:{f1_score(y_test, y_pred)}')
print(f'accuracy:{accuracy_score(y_test, y_pred)}')
```
