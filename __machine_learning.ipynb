## Supervised learning

### Regression

The most common type of ML problem. Lets say you want to predict sales based on the money spent on TV advertising, housing prices or do similar <code>**predictive analytics**</code>.<br>
First we need some collected data.



Given this data lets say you want to make a prediction for how much you could sell your property in that region where this data was collected. How learnin algorithm can help us ? Well one thing it can do is place a straight fitted line to the data and based on that make predictions. <br>

But maybe this is not the only learning algorithm we can use. Maybe its better not to fit a straight line but a <code>**quadratic function (second degree polynomial)**</code> to this data.



So the <code>**supervised learning**</code> means we gave it "right answers" on our dataset. We have mapped price with certain property in case of real estate.  And the task of the algorithm was to produce more of this "right answers" for new inputs.
It also called a <code>**regression problem, it predicts continous value outup**</code>(namely the price).

### Clasification

Looking at medical records to predict is tumor malignant or benign. If lets say we have collected data set that on horizontal axis has the sizes of tumor on vertical are these sample in <code>**one or another class**</code> (malignant or benign). We have eight example of bening and five malignant case. So if we had to predict if certain tumor case belongs to one or unfortunately to another class. ML question is can it estimate what is the probability of certain case to belong to one or another class. So this is a <code>**classification problem that deals with discrete valued output**</code> (0 or 1). Sometimes we can have more thant two possible values for the output. For example maybe there is three types of tumors. So we might try to predict a discrete value output 0,1,2,3 (sort of different classes of tumors lets say by severity).







In this graph we use only one <code>**feature(attribute)**</code> namely the size of tumor to predict whether is it malignan or benign. In real problems ussualy we have more than one feature. Lets say that we have the tumor size and the age of patient. (Malignant tumors are denoted by red circles). So our patient that has age and tumor of certain size falls somewhere in this chart. So given this data a <code>**learning algorithm might try to fit a straigth line to the data to seperate all classes**</code>. And that can predict our new patient case, hopefully he is more likely to be with benign tumor than malignant.



In this example we have two features, age and tumor size. In other ML problems we will often have more features. In this particularl example that we studied it might be:
 * Tissue thickness
 * Uniformity of cells
 * Shape of cells
 * etc.
 
Most interesting algorithms are those that can deal with not two or three features but thousands or even millions of features
that is used to make predictions. But how to deal with infinite amount of features ? How do we even store infinite number of  things in computer memory that ussualy tends to run out. It turns out that when we talk about an algorithm called <code>**[Support vector machine](https://en.wikipedia.org/wiki/Support_vector_machine)**</code> there will be a neat mathematical trick that will allow computer machine to deal with an infinite number of features.

## Unsupervised learning

Second major learning problem called unsupervised learning. In previous section we had dataset where <code>**each example was labeled either positive or negative**</code> We were explicitly told the "right answer".



In unsupervise learning we are given data set that looks likes this. It doesnt have any labels or has the same label. We are not told what to do with it what each data point is. Instead we're just told here is a data set and can you find some structure in the data?



Given this data set unsupervised algorithm might decide that data is in two different clusters and break this data into these two separate clusters. 



It's called <code>**clustering algorithm**</code> and is used in many places like [Google News](https://news.google.com/), on daily basis it looks at tens of thousand of stories and  groups them in cohesive news.



Its used in genomics, pharmatech, bioengineering. Like DNA microarray data. The idea is to group different individual and for each of them measure how much they have or do not a certain gene technically how much is it expressed. It shows degree to which idividual do or do not have a specific gene. And we can run a clustering algoritmh to group  individuals into different categories. So we do not tell algorithm in advance to which group or type these people belong to. Instead we provide bunch of data and we dont know whats in this data and can you structurize this data and cluster these individual people into types that we do not know in advance.




 * Organize server machine clusters
