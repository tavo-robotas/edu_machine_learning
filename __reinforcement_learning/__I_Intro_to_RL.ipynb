## Agent ant environment

$$H_{t} = A_{1}, O_{1}, R{1}, \ ... \,A_{t}, O_{t}, R_{t}$$

 - i.e all observable variables up to time 𝑡

In some sense this all observable variables, there might be other things hidding in the environment but agents doesn't know about those. It can't observe those things. In some sense they are irrelevent to the algorithms that we create. The algorithms can only be concerned with what the agent can actually see(experience). Remember we are trying to build that brain where the algorithm sits in that brain and we have to concerned only with things that this brain is exposed to. 

 - i.e the sensorimotor stream of a robot or embodied agent

It is easily understood on robots and they sensor data stream, but the same applys more generally like to games playing agent, trading agent. It has its own sensorimotor stream of what are the inputs that it is seeying and decision that it is able to take.  There is some well defined interface between agent and environment and we just get to see and control this stream of stuff comming into the agent.

 - What happens next depends on the this history:
     - The agent select actions

We get to build a mapping from this histories to an action. Our goal is to build this mapping (an algorithm) that picks next action from expierence.  <code>**So what happens next from the agents point of view depends on the history.**</code>

 - What happens next depends on the this history:
     - The environment selects observations/rewards

The environment looks at what it has received, at the history of what happend in terms of agent and it also looks at history and uses its history to decide what it going to do next in terms of emitting observations and rewards.

<code>**This history litteray determines the nature how things proceed, the history determins what happens next**</code><br>

## 3) Information state

An <code>**information state (a.k.a [Markov model](https://en.wikipedia.org/wiki/Markov_model))**</code>  contains all usefull information from the history.
    
<div class="alert alert-block alert-info">
<h2>Definition</h2>
    
All state S 𝑡 is <code>**Markov**</code> if and only if 
    
\begin{multline*}
\mathbb{P}[S_{t+1} | S_{t}] = \mathbb{P}[S_{t+1} | S_{1}, ..., S_{t}]
\end{multline*}

It tells us when we build state representation that constains all usefull information from the history. We are going to define a <code>**Markov property**</code> and it basically says <code>**that the probability of next state conditioned on the state your are in right now is the same as the probability of the next state if you showed all of the previous states to the system**</code> in other words we can throw away all of the previous states and just retain our current state and we will get the same characterization of future.<br>
What that does really mean ?

<code>**The future is independeant of the past givent the present**</code>

\begin{multline*}
H_{1:t} \rightarrow S_{t} \rightarrow  H_{t+1:\infty}
\end{multline*}

Whats going to happen in future if we have the <code>**Markov property**</code> then the future is independent of the past given the present. So we only need to store <code>**𝑆𝑡**</code> if we got this state representation and its <code>**Markov**</code> we can throw away the rest of the history because it doesn't give us any more information about whats going to happen in future then the state that we have. <code>**Thats Markov property and basically we can work on more compact data set and everything is going to be fine.**</code>

<code>**The state is a sufficient statistic of the future**</code> it fully charachterizes the distribution over future actions, observations, rewards. We still have to figure out what will be the optimal behaviour. All this tells us that we can make decision based on <code>**𝑆𝑡**</code> and those decisions can be optimal.

Just image the drone, that we care of the current position, linear and angular velocity, angular position would be roughly <code>**Markov state**</code> maybe wind direction as well. And if we know all these things it doesn't matter what was drone postion 5 minutes ago. That doesn't make any difference where drone will move next from now on.  Previous history is irrelevent.

<code>**In contrast**</code> we took imperfect state represtantion that is not Markov like we have just the position but not the velocity. Now were the drone is doesnt fully determine where is it going go next because we don't know how fast its moving. So we have to look back in time to figure out what actually velocity and momentum it has. 

<code>**The environment state is Markov**</code><br>

<code>**The history H is Markov**</code> just not very usefull

 - S = last 3 items in sequence ?
 - S = counts for lights, bells and levers ?
 - S = complete sequence ?

Depending of our agent state representation we can expect what will happens next. It sort like it what happens next depends on what you believe (state representation). And our job is to build an agent state that is usefull and is effective in predicting what happens next.  

## Fully Observable Environments

<code>**Full observability:**</code> agent directly observes environment state.

\begin{multline*}
O_{t} = S_{t}^{a} = S_{t}^{e}
\end{multline*}

<code>**agent state = environment state = information state**</code>

This is what we are going to work on a lot of our time.  This is the nice case , this is the type of environment where we get to see everything. Agent litteraly gets to see whats going on in the environment state (number set). Other way to say that all of these state things collapse to the same quantity. 

Formally this is a <code>**[Markov decision process (MDP)](https://en.wikipedia.org/wiki/Markov_decision_process)**</code>


##  Partially Observable Environments

<code>**Partial observability:**</code> agent indirectly observes environment.

It when agent observe <code>**indirectly**</code> environment. It doesn't get to see everything about the environment. For example a robot with a camera might not know where it is unless it has a GPS. It just get to see this camera video stream and it has to figure it out where it is within the room. <code>**It has to localize**</code>.<br>
Or a trading agent that only observes current prices. It might not know the trends or other fundamental things from where these prices are comming from.  Has to figure out what to do next based on partial information.<br>
A poker playing agent only observers public cards


<code>**Agent state ≠ environment state**</code>

We have to build an agent state that is distinct from the environment because we dont know it and maybe it doesn't exist.<br>
Formally this is a <code>**partially observable Markov decision process (POMDP)**</code>

Our job is to build this agent state <code>**(Agent must construct its onw state representation 𝑆𝑡^𝑎)**</code>. There is a lot of ways to do it but the naive way is just to rememeber everything, complete history.<br>

$$S_{t}^{a} = H_{t}$$

Or use <code>**beliefs of environment state (bayesian approach)**</code>

$$ S_{t}^{a} = (\mathbb{P}[S_{t}^{e} = s^{1}], ... , \mathbb{P}[S_{t}^{e} = s^{n}])$$

It works like, every step we build <code>**beliefs**</code>over what we think. If I don't know what is happening in the environment so i will keep a probability distribution over where i think I'm in the environment.  I will have some probability that i think the environment state is 𝑠1,  some probability the environment state is 𝑠𝑛. And this whole vector of probabilities  defines the state that we are going to use to actually decide what to do. 

<code>**Recurrent neural network**</code>
$$ S_{t}^{a} = \sigma(S_{t-1}^{a} W_{s} + O_{t}W_{o})$$

Other choices is not to use probabilities at all. And we can have any number we want inside our agent. For example recurrent neural network. That basically means that we take our agent state that we have at previous time step and we take a <code>**linear combination**</code> of the agent state that we had last time with our latest observations and that gives us a new state. So basically we have some linear transformations from our old state to the new state that takes into account of the latest observation and adding some nonlinearity around that thing and thats it. 

## Inside an RL Agent



So far we have mostly talked about the problem but not how to <code>**solve the problem**</code>.<br>

An RL agent may include one or more of these components:
 - Policy: agent’s behaviour function
 - Value function: how good is each state and/or action
 - Model: agent’s representation of the environment

We are going to be build a bit of taxonomy of how we describe reinforcement learning. In above list is the three main object that might be in our agent and they are not the only one's. Its not an exlusive list.

## Value

Next essentiall quantity is the value function. 
 - <code>**Value function is a prediction of expected future total reward**</code>
 - <code>**Used to evaluate the goodness/badness of states**</code>
 -  <code>**And therefore to select between actions, e.g**</code>
 
We need this to be able to choose between state 1 and state 2 or between action 1 and action 2 and how to do this ?
Well we want to be able to choose on correct basis and correct basis that we are trying to use here is how much total reward we expect to get in the future. 

Formally:
$$ v_{\pi}(s) = \mathbb{E}_{\pi} [R_{t} + \gamma R_{t+1}  + \gamma^{2} R_{t+2} + ... | S_{t} = s ]$$
 

 The value function depends on the way agents is behaving. So we have to index it here by π. 
 If we would have a robot which falling over a lot that robot will get very different total reward to an agent which is standing up or walking effectively. So the amount of reward we get depends on the policy. So the value function for a policy tells us how much total reward we expect to get going into the future. So we will look at expecation over the future of the reward at time 𝑡 plus the reward at the next time step 𝑡+1 and next time step. And we can also can have this discouting going into the future. This is something that says that we care more about immediate rewards then later ones.
This is the essential idea, how much reward we can get from some state onwords if we are going to follow this particular behaviour. If we are in the drone and we know the we are going to execute some particular trajectory how much reward will actually this trajectory get us. That what we try to say here. And we can compare these things then we got the basis to make good decisions. <code>**Value function is very helpfull quantinty for optimizating behaviour**</code>.

## Model

- It predict what the environment will do next
- <code>**Transitions:**</code> P predicts the next state (i.e dynamics)
- <code>**Rewards:**</code> R predicts the next (immediate) reward, e.g

So model basically is not the environment itself but sometimes it's usefull to image what will the environmnet do. To try to learn the behaviour of environment and use that model of environment to help make a plan and figure out what to do next. 

Specifically the way normally to do this is to have two parts to a model <code>**transitions**</code> and <code>**reward**</code> models. The <code>**transition model**</code> is predicting what will be the next state what will be dynamics of the environment. So if this is a drone this is the thing that says if we are in some particular position with some angle and the wind is comming from some direction that it is likely to move to this next position in another angle and so forth. Thats the dynamics of the environment and we can try to model those dynamics. We can try learn what will happen to the drone at the next step and figure that out. And we can also learn to predict how much <code>**reward**</code> we will get and so we can learn that if the drone is in some situation and its not crashing and will get one reward for staying alive.

Formally:

1)$$ P_{s\acute{s}}^{a} = \mathbb{P} [\acute{S} = \acute{s} | S = s, A = a] $$
2)$$ R_{s}^{a} = \mathbb{E} [R | S = s, A = a] $$

1. <code>**State transition model**</code> and it tells us the probability of the next state given previous state and action. 

2. <code>**Reward model**</code> and it tells us the expected reward given previous/current state and action 



Its optional to do this. We will mostly focus on model free predicition methods that don't use model at all. This is not a requirement it is not necessary to explicitly build the model of environment but it is that you can do.

## Problems with RL

### Learning and Planning

Two fundamental problems in sequential decision making

<code>**Reinforcement Learning:**</code>
 - The environment is initially unknown
 - The agent interacts with the environment
 - The agent improves its policy
 

<code>**Planning:**</code>
 - A model of the environment is known
 - The agent performs computations with its model (without any external interaction)
 - The agent improves its policy 
 - a.k.a. deliberation, reasoning, introspection, pondering, thought, search
